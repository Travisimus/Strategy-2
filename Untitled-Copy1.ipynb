```python
import pandas as pd
import pandas_ta as ta
df = pd.read_csv("BTCUSD_Candlestick_5_M_ASK_01.01.2021-01.01.2024.csv")
df["Local time"] = df["Local time"].str.replace(".000", "")
# Remove ' GMT+0100' and ' GMT+0200'
df['Local time'] = df['Local time'].str.replace(' GMT\+0100', '', regex=True)
df['Local time'] = df['Local time'].str.replace(' GMT\+0200', '', regex=True)
df['Local time'] = pd.to_datetime(df['Local time'], format='%d.%m.%Y %H:%M:%S')
df = df[df.High!=df.Low]
df.set_index("Local time", inplace = True)
df.reset_index(inplace=True)
df
```

```python
supertrend = ta.supertrend(df['High'], df['Low'], df['Close'], length=12, multiplier=3.0)
df['DEMA'] = ta.dema(df['Close'], length = 200)
# Assign the SuperTrend indicator and direction to the DataFrame
df['SUPERT_12_3.0'] = supertrend['SUPERT_12_3.0']
df['SUPERTd_12_3.0'] = supertrend['SUPERTd_12_3.0']
df['ATR'] = ta.atr(df.High, df.Low, df.Close, length = 7)
df
```

```python
def dema_signal(df, current_candle, backcandles):
    buy = 0
    sell = 0
    if (current_candle - backcandles)<=0:
        
        return 0
    else:
        for row in range(current_candle - backcandles, current_candle+1):

            if (df.Close[row]<df['DEMA'][row]):
                sell+=1

            elif (df.Close[row]>df['DEMA'][row]):
                buy+=1

            else:
                continue

        if (buy==backcandles+1):
            return 2

        elif (sell==backcandles+1):
            return 1

        else:
            return 0 
"""
count1 = 0
count2 = 0
for row in range(len(df)):
    if (dema_signal(df, row, 6)==2):
        count1+=1
    elif (dema_signal(df, row, 6)==1):
        count2+=1

print(count1, count2, df.Close[310083], df.DEMA[310082])
df
"""
```

```python
def total_signal(df, current_candle, backcandles):
    
    if dema_signal(df, current_candle, backcandles)==2 and df['SUPERTd_12_3.0'][current_candle]==1 and df['SUPERTd_12_3.0'][current_candle-1]==-1:
        
        return 2
    
    elif dema_signal(df, current_candle, backcandles)==1 and df['SUPERTd_12_3.0'][current_candle]==-1 and df['SUPERTd_12_3.0'][current_candle-1]==1:
        
        return 1
    
    else:
        
        return 0

from tqdm import tqdm
tqdm.pandas()
df['total_signal'] = df.progress_apply(lambda row: total_signal(df, row.name, 8) , axis=1) 
df
```

```python
df[df['total_signal']!=0].count()
```

```python
def SIGNAL():
    return df.total_signal
```

```python
from backtesting import Strategy
from backtesting import Backtest

class MyStrat(Strategy):
    slcoef = 1.1 #we will multipy this by ATR to get the stop loss
    TPSLRatio = 1.5
    
    def init(self):
        super().init()
        self.signal1 = self.I(SIGNAL)
        #df['RSI']=ta.rsi(df.Close, length=self.rsi_length)

    def next(self):
        super().next()
        slatr = self.slcoef*self.data.ATR[-1]
        sl = self.data['SUPERT_12_3.0'][-1]
        tp_distance = abs(self.data.Close[-1] - sl)
        TPSLRatio = self.TPSLRatio
        
        if self.signal1==2 and len(self.trades)==0:
            sl1 = sl
            tp1 = self.data.Close[-1] + slatr*TPSLRatio
            self.buy(sl=sl1, tp=tp1)
        
        elif self.signal1==1 and len(self.trades)==0:         
            sl1 = sl
            tp1 = self.data.Close[-1] - slatr*TPSLRatio
            self.sell(sl=sl1, tp=tp1)

bt = Backtest(df, MyStrat, cash = 100000)
```
